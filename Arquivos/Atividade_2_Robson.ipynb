Lista utilizada no exercício:



idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]


```python
idades = [23, 45, 31, 54, 29, 61, 40, 37, 48, 33]

print(idades)
```


1) Mostre:

a) A primeira idade

b) A última idade

c) O tamanho da lista

d) A terceira idade



```python
print("a: ", idades[0])

print("b: ", idades[-1])

print("c: Tamanho da lista: ", len(idades))

print("d: Terceira idade: ", idades[2])
```

2) Calcule manualmente:

a) A média das idades

b) A maior idade

c) A menor idade



```python
soma = 0

for idade in idades:
  soma += idade

print("Soma: ", soma)
```

```python
print("a: Média: ", soma/len(idades))
```

```python
print("b: Máx: ", max(idades))
```

```python
print("c: Min: ", min(idades))
```

3) Crie novas listas contendo:

a) Idades acima de 40

b) Idades entre 30 e 50

c) Idades pares



```python
acima_40 = [idade for idade in idades if idade > 40]
print("a: Idades acima de 40: ", acima_40)
```

```python
entre_30_e_50 = [idade for idade in idades if (idade < 50 and idade > 30)]
print("b: Idades entre 30 e 50: ", entre_30_e_50)
```

```python
idades_pares = [idade for idade in idades if idade % 2 == 0]
print("c: idades pares: ", idades_pares)
```

4) Crie uma nova lista com todas as idades aumentadas em 5 anos.



```python
lista_mais_5 = [idade+5 for idade in idades]
print("4: ", lista_mais_5)
```

5) Crie uma lista chamada grupo_risco com idades acima de 50.
Mostre quantos pacientes estão no grupo de risco.

```python
grupo_risco = [idade for idade in idades if idade > 50]
print("5: Grupo de risco: ", grupo_risco)
```

```python
print("Número de pacientes no grupo de risco: ", len(grupo_risco))
```
